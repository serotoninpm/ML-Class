```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import mglearn
```

```python
from sklearn.datasets import load_iris
```

```python
iris_dataset = load_iris()
```

```python
print("아이리스 키: {}".format(iris_dataset.keys()))
```

```python
iris_dataset.data #데이터 값
```

```python
print(iris_dataset.target) #class label
```

```python
iris_dataset.target_names #품종이름
```

```python
print(iris_dataset.DESCR) #설명
```

```python
iris_dataset.feature_names #feature이름
```

```python
print(iris_dataset.filename)#저장된 위치
```

```python
print("data의 크기 {}".format(iris_dataset['data'].shape))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'],iris_dataset['target'], test_size=0.1, random_state=13)
```

```python
y_train
```

```python
#train, test 비율 9:1인것 확인
print("X_train 크기: {}".format(X_train.shape))
print("y_train 크기: {}".format(y_train.shape))
print("X_test 크기: {}".format(X_test.shape))
print("y_test 크기: {}".format(y_test.shape))
```

```python
%matplotlib inline

iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)

#산점도 그리기, bins는 막대기 개수, c는 color
pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15,15), marker="o", hist_kwds={'bins':20}, s=60, alpha=.8, cmap=mglearn.cm3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, y_train)
```

```python
X_random = np.array([[5,2.9,1,0.2]])
```

```python
print(X_random)
```

```python
prediction = knn.predict(X_random)
```

```python
print(prediction)
print(iris_dataset['target_names'][prediction])
```

```python
y_pred =knn.predict(X_test)
print("테스트 세트에 대한 예측값\n",y_pred)
```

```python
print("테스트세트의 정확도: {:.2f}".format(np.mean(y_pred == y_test)))
```

```python
print("테스트세트의 정확도: {:.2f}".format(knn.score(X_test, y_test)))
```
