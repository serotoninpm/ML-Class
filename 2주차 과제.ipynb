# Scikkit-learn knn

```python
from sklearn.model_selection import train_test_split
X, y = mglearn.datasets.make_forge()
```

```python
print("X.shape:{}".format(X.shape), "y.shape: {}".format(y.shape))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=0)
```

```python
print("X_train.shape:{}".format(X_train.shape), "X_test.shape: {}".format(X_test.shape))
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=2)
```

```python
clf.fit(X_train, y_train) # 2개로 파라미터 변경
```

```python
print("테스트 세트 예측: {}".format(clf.predict(X_test)))
```

```python
print("테스트 세트 정확도: {:.2f}".format(clf.score(X_test, y_test)))
```

```python
mglearn.plots.plot_knn_classification(n_neighbors=1)
```

```python
mglearn.plots.plot_knn_classification(n_neighbors=3)
```

```python
fig, axes = plt.subplots(1,3, figsize=(10,3))
for n_neighbors, ax in zip([1,3,9], axes):
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X,y)
    mglearn.plots.plot_2d_separator(clf, X, fill= True, eps = 0.5, ax= ax, alpha=.4)
    mglearn.discrete_scatter(X[:,0], X[:,1],y,ax=ax)
    ax.set_title("{} 이웃".format(n_neighbors))
    ax.set_xlabel("특성 0")
    ax.set_ylabel("특성 1")
axes[0].legend(loc=3)
```

# Wisconsin Breast Cancer Data

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
print(cancer)
```

```python
print("cancer.keys()\n", cancer.keys())
```

```python
cancer.data
```

```python
print("데이터의 형태", cancer.data.shape)
```

```python
print(cancer.target_names)
```

```python
print(cancer.target)
```

```python
#0부터 올림 차순으로 정렬한 뒤, count
np.bincount(cancer.target)
```

```python
print("클래스별 샘플 갯수:\n", {n : v for n, v in zip(cancer.target_names, np.bincount(cancer.target))})
```

```python
print("특성 이름\n", cancer.feature_names)
```

```python
print("cnacer.data: \n{}".format(cancer.data))
```

```python
print("cnacer.target: \n{}".format(cancer.target))
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify = cancer.target, random_state=66)
# stratify는 한쪽에 class label이 쏠리지 않게 조정해줌
```

```python
training_accuracy =[]
test_accuracy =[]

# 1부터 10까지 neighbor지정
neighbor_settings = range(1,11)

for n_neighbors in neighbor_settings:
    # 모델 생성
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    # 훈련세트 정확도 저장
    training_accuracy.append(clf.score(X_train,y_train))
    # 일반화 정확도 저장
    test_accuracy.append(clf.score(X_test, y_test))
    
plt.plot(neighbor_settings, training_accuracy, label="훈련 정확도")
plt.plot(neighbor_settings, test_accuracy, label="테스트 정확도")
plt.ylabel("정확도")
plt.xlabel("n_neighbors")
plt.legend()
```
